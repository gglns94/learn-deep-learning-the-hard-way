### Computing the nullspace

Turn definition to algorithm.

#### Example

$$
A=
\begin{bmatrix} 
1 & 2 & 2 & 2 \\
2 & 4 & 6 & 8 \\
3 & 6 & 8 & 10
\end{bmatrix}
$$

The process of elimination does not change nullspace.

$$
\rightarrow \begin{bmatrix}
\underline{1} & 2 & 2 & 2\\
0 & 0 & 2 & 4\\
0 & 0 & 2 & 4\\
\end{bmatrix}$$

$$
\rightarrow \begin{bmatrix}
\underline{1} & 2 & 2 & 2\\
0 & 0 & \underline{2} & 4\\
0 & 0 & 0 & 0\\
\end{bmatrix}
$$

$$=U$$

**Rank** of $A$ = # of pivots (in this case 2)

### Pivot variables - free variables

- Columns containing pivots : Pivot Columns
- Others : Free Columns

Do back substitution for above example.

Why can I call these free columns ?
Because we can assign any value for elements multipied to those columns.
So in above example, we could do following :
    
$$
x = \begin{bmatrix}
? \\ 
1 \\ 
? \\
0 \\
\end{bmatrix}
$$

What does this mean? We can express the equation like following :

$x_1 + 2x_0 + 2x_3 + 2x_4 = 0$<br>
$2x_3 + 4x_4 = 0$

$\rightarrow$

$x_1 + 2 + 2x_3 = 0$<br>
$2x_3 = 0$<br>

Now we can find the solution : 
$
x = \begin{bmatrix}
-2 \\
1 \\
0 \\
0
\end{bmatrix}
$

What if we assign different values : 
$
x = \begin{bmatrix}
? \\
0 \\
? \\
1 \\
\end{bmatrix}
$

Then, the solution is :
$
x = \begin{bmatrix}
2 \\
0 \\
-2 \\
1 \\
\end{bmatrix}
$

$$
x = 
c \begin{bmatrix}
-2 \\
1 \\
0 \\
0
\end{bmatrix}
+ d \begin{bmatrix}
2 \\
0 \\
-2 \\
1 \\
\end{bmatrix}
$$

Rank : $r = 2$<br>
Free columns : $n - r = 4 -2$<br>

### RREF

One more step to make the $U$ better

$R =$ <u>reduced</u> row echelon form<br>
zeros above + below <u>pivots</u> = 1

$$
\begin{bmatrix}
1 & 2 & 2 & 2 \\
0 & 0 & 2 & 4 \\
0 & 0 & 0 & 0 \\
\end{bmatrix}
$$

$$
\rightarrow
\begin{bmatrix}
1 & 2 & 0 & -2 \\
0 & 0 & 2 & 4 \\
0 & 0 & 0 & 0 \\
\end{bmatrix}
$$

$$
\rightarrow
\begin{bmatrix}
\underline{1} & 2 & 0 & -2 \\
0 & 0 & \underline{1} & 2 \\
0 & 0 & 0 & 0 \\
\end{bmatrix}
$$

$$= R = rref(A)$$

It's got all the information as clear as can be.
- Notice $I$ in privot rows/cols

**rref form** 
$$
R = \begin{bmatrix}
I & F \\ 
0 & 0 \\
\end{bmatrix}
$$

$r$ pivot rows,
$r$ pivot cols, 
$n-r$ free cols

#### Nullspace matrix

$RN = 0$

$$
N = \begin{bmatrix}
-F \\
N \\
\end{bmatrix}
$$

Cols of $N$ are special solutions.

$$
Rx = 0
\rightarrow \begin{bmatrix}I & F \\\end{bmatrix} \begin{bmatrix} x_{pivot} \\ x_{free} \\ \end{bmatrix}
$$

Another example :
    
$$
A = \begin{bmatrix}
1 & 2 & 3 \\
2 & 4 & 6 \\
2 & 6 & 8 \\
2 & 8 & 10 \\
\end{bmatrix}
\rightarrow
\begin{bmatrix}
\underline{1} & 2 & 3 \\
0 & 0 & 0 \\
0 & 2 & 2 \\
0 & 4 & 4 \\
\end{bmatrix}
\rightarrow
\begin{bmatrix}
\underline{1} & 2 & 3 \\
0 & \underline{2} & 2 \\
0 & 0 & 0 \\
0 & 0 & 0 \\
\end{bmatrix}=U
(rank = 2, free = 1)
$$

$$
R = \begin{bmatrix}
\underline{1} & 0 & 1 \\
0 & \underline{1} & 1 \\
0 & 0 & 0 \\
0 & 0 & 0 \\
\end{bmatrix}
$$
